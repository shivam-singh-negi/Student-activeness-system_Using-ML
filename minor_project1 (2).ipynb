```python
import pandas as pd
import numpy as np
import random 
from google.colab import files
```

```python
 
import pandas as pd
 
# initialise data of lists.
a = [0,1, 2, 3,4,5,6]
b= [0,1, 2, 3,4,5,6]
c = [0,1, 2, 3,4,5,6]

cd= pd.MultiIndex.from_product([a, b,c], names = ["missed", "wrong","right"])
df=cd.to_frame(index = False)
print(df)
print(len(df))
```

```python
df["Sum"]=df["missed"]+df["wrong"]+df["right"]
df.drop(df[df['Sum']!=6].index, inplace = True)
print(df)
print(len(df))
```

```python
df["Accur"]=(df["right"]/(df["right"]+df["wrong"]))*100
df["Accur"].fillna(value = 0,
          inplace = True)
print(df)
```

```python
df["missed_factor"]=(df["missed"]/(df["missed"]+(6-df["missed"])))*100
print(df)
```

```python
df["Tot_clicks"]= np.random.randint(1,100, size=len(df))
df["Pop_clicks"]=np.random.randint(1,df["Tot_clicks"], size=len(df))
df["effective_clicks"]=(df["Pop_clicks"]/(df["Tot_clicks"]))*100
print(df)
```

```python
df["Tot_scrolls"]= np.random.randint(1,100, size=len(df))
df["Pop_scrolls"]=np.random.randint(1,df["Tot_scrolls"], size=len(df))
df["effective_scrolls"]=(df["Pop_scrolls"]/(df["Tot_scrolls"]))*100
print(df)
```

```python
df1=df[["Accur","missed_factor","effective_clicks","effective_scrolls"]].copy()
print(df1)
```

```python
tempa=[0,1,2,3,4,5,6]
tempb=[0,1,2,3,4,5,6]
tempc=[0,1,2,3,4,5,6]
temp_e_click = df["effective_clicks"].drop_duplicates()
temp_e_scroll= df["effective_scrolls"].drop_duplicates()
cd3= pd.MultiIndex.from_product([tempa,tempb,tempc,temp_e_click,temp_e_scroll], names = ["wrong","right","missed","temp_e_click","temp_e_scroll"])
cd3
```

```python
df4=cd3.to_frame(index = False)
print(df4)
print(len(df4))
```

```python
df4["Sum"]=df4["missed"]+df4["wrong"]+df4["right"]
df4.drop(df4[df4['Sum']!=6].index, inplace = True)
print(df4)
print(len(df4))
```

```python
df4["Accur"]=(df4["right"]/(df4["right"]+df4["wrong"]))*100
df4["Accur"].fillna(value = 0,
          inplace = True)
print(df4)
```

```python
df4["missed_factor"]=(df4["missed"]/(df4["missed"]+(6-df4["missed"])))*100

print(df4)
```

```python
df4.reset_index(inplace = True, drop = True)
df4.drop_duplicates(keep=False,inplace=True)
print(df4)
```

```python
fe=df4[["Accur","missed_factor","temp_e_click" , "temp_e_scroll" ]]
print(fe)
```

```python
fe["w_Accur"]=(fe["Accur"]/10)
print(fe)
```

```python
fe.tail(100)
```

```python
fe["w_missed"]=(100-fe["missed_factor"])*0.3
print(fe["w_missed"])
```

```python
fe["w_click"]=fe["temp_e_click"]*0.3
print(fe["w_click"])
```

```python
fe["w_scroll"]=fe["temp_e_scroll"]*0.3
print(fe["w_scroll"])
```

```python
print(fe)
```

```python
fe["sum"]=fe["w_Accur"]+fe["w_missed"]+fe["w_click"]+fe["w_scroll"]
print(fe["sum"])
```

```python
def assignlabel(n):
  if n>=75.0:
    return "Highly Active"
  elif n>=50.0:
    return "Active"
  elif n>=25.0:
    return "Less Active"
  else:
    return "Least Active"
```

```python
fe["Result"]=fe["sum"].apply(assignlabel)
print(fe)
```
